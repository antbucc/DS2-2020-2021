```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
MEAN = 30
VAR = 5

NUM_VALUES = 2000
```

```python
data = [np.random.normal(MEAN, VAR) for i in range(NUM_VALUES)]
data
```

```python
NUM_BINS = 500

plt.hist(np.cumsum(data), bins=NUM_BINS, label='Instances per {:.2f}s'.format(sum(data)/NUM_BINS))
plt.legend()
plt.show()
```
