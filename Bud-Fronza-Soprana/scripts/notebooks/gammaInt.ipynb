```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
import scipy.interpolate

from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
mean = 2
var = 5

scale = var/(mean-1)
alpha = (mean-1)/scale

g = gamma(alpha, 0, scale)

num=30
```

```python
x = np.linspace(0,num,num=num+1)
y = [g.cdf(i+0.5)-g.cdf(i-0.5) for i in x]

x = x+0.5

fig, ax = plt.subplots()
ax.set_xlim([0.5, num])

plt.hist(x, weights=y, label='Non cumulative (num. identities = x)', bins=num)

x = x+0.5

plt.plot(x, np.cumsum(y), label='Cumulative (num. identities <=x)', marker='x', linewidth=1)

ax.xaxis.set_major_locator(MultipleLocator(3))
ax.xaxis.set_minor_locator(MultipleLocator(1))

ax.set_ylabel('Probabilty')
ax.set_xlabel('Number of itentities')
plt.legend()
plt.grid()
plt.show()
```
